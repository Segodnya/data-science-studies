# Первичная обработка данных

```python
# приведем название столбцов к единому виду
data.rename(
    columns={'studytime, granular': 'studytime_granular'}, inplace=True)
```

```python
# сформируем списки столбцов с количественными и номинативными переменными
# формировать списки будем по типу данных
# готовые списки будут полезны при применении методов анализа для количественных и номинативных признаков

list_numeric = []
list_nominative = []

for elem in data.columns:
    if data[elem].dtype == 'O':
        list_nominative.append(elem)
    else:
        list_numeric.append(elem)

# check
print('numeric = ', list_numeric)
print()
print('nominative = ', list_nominative)
print()
```

```python
# избавимся от NaN в столбцах с количественными переменными
# data = data.dropna(subset=list_numeric) - применение метода нецелесообразно,
# т.к. мы теряем более ста наблюдений (~треть датасета)

# опираясь на сделанные выше базовые расчеты статистик,
# выберем медиану как наиболее подходящую замену для пропущенных значений
for col in list_numeric:
    data[col] = data[col].fillna(data[col].median())

# check
data.describe()
```

Теперь датасет не содержит пропущенныз значений в количественных признаках. Можем переходить к их анализу.

```python
# заменим NaN на unknown в столбцах с номинативными переменными
data = data.fillna(value='unknown')
```

Удаление или замена пустых значений для номинативных признаков не выглядит хорошей идеей - лучше воспользоваться фиктивным значением - возможно, при дальнейшем анализе удастся выявить влияние отсутствия данных по какому-то из параметров на итоговый результат теста по математике.

# Тестирование номинативных переменных

Протестируем два набора данных: data_1 - очищенный датасет от учеников, проваливших экзамен и набравших 0 баллов, и data - исходный датасет, содержащий всех учеников. Предварительная обработка data, как и data_1, проведена на этапе предобработки данных.

```python
# определим функция для т-тестирования номинативных признаков
# тестирование проводим в отношении к результирующей переменной score
def get_stat_dif(column, df):
    cols = df.loc[:, column].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(df.loc[df.loc[:, column] == comb[0], 'score'],
                     df.loc[df.loc[:, column] == comb[1], 'score']).pvalue \
                <= 0.1/len(combinations_all):  # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
```

```python
# протестируем датасет data_1, очищенный от учеников, набравших на экзамене 0 баллов
for col in ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason'
            , 'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher'
            , 'internet', 'romantic']:
    get_stat_dif(col, data_1)
```

```python
# протестируем датасет data, содержащий в том числе учеников, набравших на экзамене 0 баллов
for col in ['school', 'sex', 'address', 'famsize', 'Pstatus'
           , 'Mjob', 'Fjob', 'reason', 'guardian', 'schoolsup'
           , 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']:
    get_stat_dif(col, data)
```

## Краткий вывод по тестированию номинативных переменных
